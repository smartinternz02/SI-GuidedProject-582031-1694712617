Univariate analysis

since there are no categorical data we cannot use pie chart


```python
#Histogram
plt.hist(df['number of floors'])
plt.show()
```

```python
sns.displot(df['Number of schools nearby'])
plt.show()
```

```python
#barplot
plt.bar(df.index,df['Area of the basement'])
plt.show()
```

BIVARIATE ANALYSIS


```python
#scatter plot
plt.scatter(df['Number of schools nearby'],df['Price'])
plt.ylabel("Price")
plt.xlabel("Number of schools nearby")
plt.show()
```

```python
sns.jointplot(x=df['Number of schools nearby'],y=df['Price'])
plt.show()
```

MULTIVARIATE ANALYSIS


```python
sns.heatmap(df.corr())
plt.show()
```

```python
sns.pairplot(df,vars=['number of bedrooms','number of bathrooms','lot area','number of floors','Area of the basement'])
plt.show()
```

```python
#if we want to use only a few variables in pairplot we can do vars=[<list of the variable names>]
```
